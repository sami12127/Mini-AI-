## Opgave 1b

Bekijk onderstaande code. Stel we hebben een geheugen van een list van 1000 elementen.

- Wat gebeurt er met geheugen als je 10_000 of 100_000 items gebruikt?

        De "_" word weg gelaten en de items worden gewoon gezien als 10000 of 100000

- Waarom groeit geheugen op deze manier?

        Omdat er meer elementen in de list zitten, dus is er meer geheugen nodig.

```python
lst = []
for i in range(10_000):
    lst.append(i)

print("Aantal elementen:", len(lst))
print("Geheugen (bytes) voor 10_000 elementen:", sys.getsizeof(lst))

lst2 = []
for i in range(100_000):
    lst2.append(i)

print("Aantal elementen:", len(lst2))
print("Geheugen (bytes) voor 100_000 elementen:", sys.getsizeof(lst2))
```

## Opgave 2
### Vergelijk append vs insert(0)

- Welke is langzamer? Waarom?

        Bij Append word een element aan het einde van de lijst toegevoegd (O(1)), en bij insert word een element ingevoerg op index "i", dus alle elementen die erna komen moeten opschuiven. (O(n)). daarom is insert langzamer.

- Wat leert dit over Lijsten en inserts?

        Dat python lijsten inefficiënt zijn voor het toevoegen van een element aan het begin of midden van een lijst (insert).

```python
n = 100_000

lijst = []
start = time.time()
for i in range(n):
    lijst.append(i)
print("Tijd voor append:", time.time() - start)

lijst = []
start = time.time()
for i in range(n):
    lijst.insert(0, i)
print("Tijd voor insert(0, i):", time.time() - start)
```
