## Opgave 2
### Vergelijk append vs insert(0)

- Welke is langzamer? Waarom?

Als je de code runt zie je dat Insert aanzienelijk langzamer is: insert(0) → voegt iets toe vooraan → alles moet opschuiven → traag vergelijken met iets achteraan de lijst de zetten zonder iets te verschuiven

- Wat leert dit over Lijsten en inserts?

Dat inserts langzamer is omdat het moet verschuiven ipv iets achteraan de lijst toe te voegen bij append.

```python
# append testen
lijst = []
start = time.time()

for i in range(100000):
    lijst.append(i)

print("Append tijd:", time.time() - start)


# insert(0) testen
lijst = []
start = time.time()

for i in range(100000):
    lijst.insert(0, i)

print("Insert(0) tijd:", time.time() - start)
```
