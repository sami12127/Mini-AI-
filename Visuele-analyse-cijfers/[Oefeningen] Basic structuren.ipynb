## Opgave 2
### Vergelijk append vs insert(0)

- Welke is langzamer? Waarom?

Als je de code runt zie je dat Insert aanzienelijk langzamer is: insert(0) → voegt iets toe vooraan → alles moet opschuiven → traag vergelijken met iets achteraan de lijst de zetten zonder iets te verschuiven

- Wat leert dit over Lijsten en inserts?

Dat inserts langzamer is omdat het moet verschuiven ipv iets achteraan de lijst toe te voegen bij append.

```python
# append testen
lijst = []
start = time.time()

for i in range(100000):
    lijst.append(i)

print("Append tijd:", time.time() - start)


# insert(0) testen
lijst = []
start = time.time()

for i in range(100000):
    lijst.insert(0, i)

print("Insert(0) tijd:", time.time() - start)
```

### Set lookup vs list lookup

Maak een lijst en een set van 1 miljoen elementen.
- Hoe lang duurt het om laatste element op te zoeken in elk?
- En voor een hoger getal?

```python
# Maak lijst en set van 1 miljoen elementen
N = 1_000_000
my_list = list(range(N))
my_set = set(range(N))

# Laatste element opzoeken
start = time.time()
last_in_list = my_list[-1]
end = time.time()
print(f"Lijst laatste element: {last_in_list}, tijd: {(end-start)*1e6:.2f} microseconden")

start = time.time()
last_in_set = (N-1 in my_set)
end = time.time()
print(f"Set laatste element aanwezig? {last_in_set}, tijd: {(end-start)*1e6:.2f} microseconden")

# Hoger getal opzoeken (niet aanwezig)
high_value = N + 1000

start = time.time()
found_in_list = high_value in my_list
end = time.time()
print(f"Hoger getal in lijst aanwezig? {found_in_list}, tijd: {(end-start):.4f} s")

start = time.time()
found_in_set = high_value in my_set
end = time.time()
print(f"Hoger getal in set aanwezig? {found_in_set}, tijd: {(end-start)*1e6:.2f} microseconden")
```

## Opgave 6

- Hoe kun je een stack simuleren met python?

In Python kun je een stack heel eenvoudig simuleren met een list, omdat een list append() en pop()
- voeg 200000 elementen toe. Haal ze er vervolgens weer af. Hoeveel tijd heeft het gekost?

Zie code

```python
stack = []

N = 200_000

start_push = time.time()
for i in range(N):
    stack.append(i)
end_push = time.time()

print(f"Tijd om {N} elementen toe te voegen (push): {(end_push - start_push):.4f} seconden")


start_pop = time.time()
while stack:
    stack.pop()
end_pop = time.time()

print(f"Tijd om {N} elementen te verwijderen (pop): {(end_pop - start_pop):.4f} seconden")
```

## Opgave 7

- Hoe kun je een queue simuleren met python?

Met een lijst (list) door append() te gebruiken om toe te voegen en pop(0) om het eerste element te verwijderen.
- Voeg 200000 elementen toe. Haal ze er weer uit. Hoe lang duurt het?

Dat duurt relatief lang, omdat pop(0) traag is (alle elementen moeten opschuiven). De tijd loopt merkbaar op.
- Doe nu hetzelfde voor dequeue. Je kunt deque() gebruiken. Hoe lang duurt dat? Wat valt je op? 

Dat gaat veel sneller, deque.popleft() is efficiënt (O(1)), terwijl list.pop(0) traag is (O(n)).


```python
queue = []
start = time.time()

# 200000 elementen toevoegen
for i in range(200000):
    queue.append(i)

# 200000 elementen verwijderen
for i in range(200000):
    queue.pop(0)

end = time.time()
print("Tijd met list:", end - start)
```

```python
queue = deque()
start = time.time()

# 200000 elementen toevoegen
for i in range(200000):
    queue.append(i)

# 200000 elementen verwijderen
for i in range(200000):
    queue.popleft()

end = time.time()
print("Tijd met deque:", end - start)
```

## Opgave 8

- Python heeft geen ingebouwde linked list — bouw een linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next

        current.next = new_node

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


# Test
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)

ll.print_list()
```

- Waarom is append O(n) in deze implementatie?

Append is O(n) omdat je telkens vanaf de head door de hele lijst moet lopen om het laatste element te vinden. Als er al n elementen zijn, kost dat n stappen. Daarom is de tijdcomplexiteit O(n).
