### Inleiding

Een fictief bedrijf heeft een apparaat ontwikkeld dat MysteryDevice heet. Het apparaat heeft een klein schermpje waar de gebruiker een handgeschreven getal kan invoeren met een stylo pen. Het getal moet worden omgezet naar een digitale representatie van getal en doorgegeven aan een gekoppeld apparaat.

Je doel is om een model te trainen dat efficiënt draait op dit apparaat, waarbij het geheugenverbruik en de rekentijd zo klein mogelijk worden gehouden.

Je ontwikkelt een AI-systeem voor de MysteryDevice. Ontwerp een volledig werkend systeem dat een handgeschreven cijfer herkent, binnen deze (hardware) constraints:

- 256 KB RAM
- 1 MB opslag (beschikbaar voor je programma)
- geen GPU
- embedded Python
- Numpy pre-installed
- gebruiker tekent zelf cijfers op een touchscreen

Je mag aannemen dat het werkelijk schrijven van een getal op het schermpje al werkt en is geimplementeerd. Daarbij wordt elke keer een plaatje van 28 bij 28 pixels aangemaakt als een **.png** bestand (andere extensie mag ook). Ga er van uit dat het een kleurenafbeeding (RGB) is die wel naar grijswaardes moet worden omgezet.

Je kunt er ook van uitgaan dat je je model op een ander apparaat mag trainen, zodat je geen rekening hoeft te houden met MNIST data (grotte), Mystery Device heeft immers geen netwerkverbinding om MNIST te downloaden.

***Je mag aannemen dat je ook geen rekening hoeft te houden met RAM voor Python en Numpy libraries zelf, alleen voor je eigen model.***

***Je mag aannemen dat er dat je de ruimte die de libraries Python en Numpy innemen buiten beschouwing kunt laten. De 1MB opslag is dus beschikaar voor je.***

***We gebruiken de bekende MNIST (https://en.wikipedia.org/wiki/MNIST_database) database voor het trainen van ons algoritme.***

- Schrijf duidelijke commentaar in je code
- Indien gevraagd moet je de code kunnen uitleggen
