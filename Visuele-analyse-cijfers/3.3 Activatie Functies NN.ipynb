### Taak 1: Bereken de output van een layer met ReLU activatie

We hebben een layer met 4 nodes. De vorige layer geeft de volgende inputs:

```python
x = [1.5, -2.0, 3.0, 0.5]
```

De biasvector is 

b = [0.1, -0.2, 0.3, 0.4]

De gewichtenmatrix (voor elke node in deze layer) is als volgt:

```python
# Elke rij hoort bij een node in de huidige layer
W = [
    [0.2, -0.5, 0.3, 0.1],   # Node 1
    [-0.3, 0.8, -0.2, 0.4],  # Node 2
    [0.5, -0.1, 0.6, -0.2],  # Node 3
    [-0.4, 0.2, 0.1, 0.3]    # Node 4
]
```

```python
import numpy as np

x = [1.5, -2.0, 3.0, 0.5]
b = [0.1, -0.2, 0.3, 0.4]
W = [
    [0.2, -0.5, 0.3, 0.1],
    [-0.3, 0.8, -0.2, 0.4],
    [0.5, -0.1, 0.6, -0.2],
    [-0.4, 0.2, 0.1, 0.3]
]

z = np.dot(W, x) + b
y = np.maximum(0, z)

print("z (pre-activatie):", z)
print("y (na ReLU):", y)
```

### Taak 2 Sigmoid

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

print(sigmoid(3))
print(sigmoid(100))
print(sigmoid(10000))
print(sigmoid(-1))
print(sigmoid(0))

# Sigmoid wordt vaak gebruikt omdat de output altijd tussen 0 en 1 ligt.
# Dit maakt het geschikt als waarschijnlijkheid bij binaire classificatie.
```

### Taak 3 Softmax

Gegeven: z = [3.0, 1.0, 0.2]

```python
import math

def softmax(z):
    e_z = [math.exp(x) for x in z]
    som = sum(e_z)
    return [v / som for v in e_z]

z1 = [3.0, 1.0, 0.2]
resultaat1 = softmax(z1)
print("Softmax van", z1, ":", resultaat1)
print("Som:", sum(resultaat1))
print("Hoogste waarschijnlijkheid bij klasse:", resultaat1.index(max(resultaat1)))

z2 = [0.1, 0.1, 0.1]
resultaat2 = softmax(z2)
print("\nSoftmax van", z2, ":", resultaat2)
print("Som:", sum(resultaat2))
```

## Taak 4

**Kat / geen kat (binaire classificatie):**
- Activatiefunctie: Sigmoid
- Sigmoid geeft een waarde tussen 0 en 1, wat je kunt interpreteren als een kans (bv. 0.8 = 80% kans dat het een kat is).
- ReLU zou niet werken: het geeft geen kansen terug en heeft geen bovengrens, waardoor je de output niet kunt interpreteren als waarschijnlijkheid.

**Cijfer herkennen (0–9, multi-class):**
- Activatiefunctie: Softmax
- Softmax geeft voor elke klasse een kans, en de kansen tellen op tot 1. Zo kun je zien welk cijfer het meest waarschijnlijk is.
- ReLU geeft geen kansen en zorgt niet dat de outputs optellen tot 1.

**Spam / geen spam (binaire classificatie):**
- Activatiefunctie: Sigmoid
- Zelfde redenering als bij kat/geen kat: het is een binair probleem, en Sigmoid geeft een kans tussen 0 en 1.
- ReLU zou outputs kunnen geven groter dan 1, wat niet bruikbaar is als kans.

## Taak 5

**Waarom de redenering deels klopt:**
Bij een classificatieprobleem zijn negatieve outputwaarden inderdaad niet zinvol als kans. ReLU verwijdert negatieve waarden, dus in dat opzicht lijkt het logisch.

**Waarom het toch een slecht idee is:**
ReLU heeft geen bovengrens. De output kan elke positieve waarde aannemen (bv. 5, 100, 1000). Dit zijn geen kansen. Je kunt de outputs niet vergelijken of interpreteren als "hoe zeker is het model?". Bovendien werkt de loss-functie (zoals cross-entropy) niet correct met ReLU-outputs.

**Beter alternatief:**
- Sigmoid voor binaire classificatie: output altijd tussen 0 en 1, direct te interpreteren als kans.
- Softmax voor multi-class classificatie: outputs tellen op tot 1, elke output is een kans per klasse.

## Taak 6

**Zelfbedachte waarden:**
- x1 = 1, x2 = 2
- w1 = 0.5, w2 = -1.0
- b = -0.5

**Berekening:**
z = w1 * x1 + w2 * x2 + b
z = 0.5 * 1 + (-1.0) * 2 + (-0.5)
z = 0.5 - 2.0 - 0.5
z = -2.0

ReLU(-2.0) = max(0, -2.0) = **0**

**Waarom ReLU hier nuttig is:**
ReLU onderdrukt deze neuron volledig. De neuron 'vuurt niet'. Dit zorgt voor sparse activatie: niet alle neuronen zijn actief tegelijk, wat het netwerk efficiënter maakt en overfitting helpt voorkomen.

## Taak 7

**Kan de output na ReLU nul zijn?**

Ja, dat kan.

**Wanneer wel:**
Als de negatieve bias groot genoeg is om de som van alle positieve gewichten maal positieve inputs te overtreffen. Dus als: w1*x1 + w2*x2 + ... + b < 0, dan is de output na ReLU = 0.

**Wanneer niet:**
Als de gewogen som van de inputs groter is dan de absolute waarde van de bias. Dan is de pre-activatie positief, en ReLU geeft die waarde gewoon terug.

## Taak 8

**Wat de output betekent:**
De step-functie geeft 0 of 1 terug. 0 betekent 'nee / klasse A', 1 betekent 'ja / klasse B'. Dit is een harde beslissing zonder tussenwaardes.

**Waarom accuracy logisch is:**
Bij een binair resultaat (0 of 1) kun je simpel tellen hoeveel voorspellingen correct zijn. Dat is precies wat accuracy meet: aantal goed / totaal.

**Wanneer accuracy misleidend kan zijn:**
Als de dataset sterk ongebalanceerd is. Bijvoorbeeld: 95% van de data is klasse 0. Een model dat altijd 0 voorspelt haalt dan 95% accuracy, maar is volledig nutteloos voor het herkennen van klasse 1. In dat geval zijn precision, recall of F1-score betere metrics.

## Taak 9

**Pre-activatie als rekensom:**
z = w1*x1 + w2*x2 + b
z = 0.5*2 + 3*(-1) + (-1)
z = 1 - 3 - 1
z = **-3**

**Waarom we deze stap apart bekijken:**
De pre-activatie (z) is de gewogen som van de inputs. Dit is een lineaire berekening. We bekijken dit apart omdat de activatiefunctie daarna een niet-lineaire transformatie toepast. Zo zie je duidelijk wat het netwerk 'berekent' vóór de beslissing.

**Wat er zou gebeuren zonder activatiefunctie:**
Zonder activatiefunctie is het netwerk puur lineair, ongeacht hoeveel lagen het heeft. Meerdere lineaire lagen samen zijn nog steeds één lineaire functie. Het netwerk kan dan geen complexe, niet-lineaire patronen leren.

## Taak 10

**Wat zijn logits:**
Logits zijn de ruwe, onbewerkte outputwaarden van de laatste laag van een neuraal netwerk, voordat er een activatiefunctie op is toegepast. Ze hebben geen vaste schaal en kunnen negatief, nul of positief zijn.

**Waarom zijn logits geen waarschijnlijkheden:**
Logits kunnen elke waarde aannemen. Ze tellen niet op tot 1 en kunnen negatief zijn. Waarschijnlijkheden moeten altijd tussen 0 en 1 liggen en optellen tot 1.

**Wat doet Softmax conceptueel:**
Softmax zet de logits om naar waarschijnlijkheden. Het vergelijkt de waarden ten opzichte van elkaar: hogere logits krijgen een hogere kans. De outputs tellen altijd op tot 1, zodat je ze kunt interpreteren als 'hoe zeker is het model voor elke klasse'.
