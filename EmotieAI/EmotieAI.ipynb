# Emotie AI

Een simpel programma dat een paar emoties herkent op basis van woorden in de tekst.

Het werkt zo: voor elke emotie is er een lijstje met woorden. Komt zo'n woord
in de tekst voor, dan is dat de emotie. Anders is het neutraal.

## Stap 1: de woordenlijsten

```python
blij = ["blij", "leuk", "fijn", "happy", "gelukkig", "top"]
boos = ["boos", "irritant", "haat", "stom", "kwaad"]
verdrietig = ["verdrietig", "huilen", "down", "alleen", "rot"]
moe = ["moe", "slaap", "uitgeput", "bekaf"]
bang = ["bang", "angst", "eng", "nerveus"]
```

## Stap 2: de functie

Deze functie kijkt of er een woord uit een lijstje in de tekst staat.
Zodra hij iets vindt, geeft hij die emotie terug.

```python
def vind_emotie(tekst):
    tekst = tekst.lower()

    for woord in blij:
        if woord in tekst:
            return "blij"
    for woord in boos:
        if woord in tekst:
            return "boos"
    for woord in verdrietig:
        if woord in tekst:
            return "verdrietig"
    for woord in moe:
        if woord in tekst:
            return "moe"
    for woord in bang:
        if woord in tekst:
            return "bang"

    return "neutraal"
```

## Stap 3: het programma

Typ iets in. Typ 'stop' om te stoppen.

```python
print("Emotie AI - typ iets en ik raad je emotie (typ 'stop' om te stoppen)")

while True:
    tekst = input("> ")

    if tekst == "stop":
        print("Tot ziens!")
        break

    emotie = vind_emotie(tekst)
    print("Ik denk dat je je", emotie, "voelt.")
```
